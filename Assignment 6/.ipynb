```python
x = data['YEAR']
y = data['ANNUAL']
```

```python
plt.plot(x, y, 'o', color='magenta')  # Change 'red' to any color you prefer
plt.xlabel('Year')
plt.ylabel('Annual Temperature')
plt.title('Annual Temperature over Years')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

print(f"x training dataset: {x_train.shape}")
print(f"x testing dataset: {x_test.shape}")
print(f"y training dataset: {y_train.shape}")
print(f"y testing dataset: {y_test.shape}")
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#y = w.x + b

model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
plt.scatter(x_test, y_test, color='magenta')
plt.plot(x_test, y_pred, color='black', linewidth=3)
plt.xlabel('Year')
plt.ylabel('Annual Temperature')
plt.title('Annual Temperature over Years')
plt.show()
```

```python
sns.regplot(data=df, x=x_train, y=y_train, color='pink', line_kws={"color": "red", "alpha": 0.7, "lw": 2})
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"MSE:  {mean_squared_error(y_test,y_pred)}")
print(f"MAE:  {mean_absolute_error(y_test,y_pred)}")
print(f"R-Sqaure :  {r2_score(y_test,y_pred)}")
```
