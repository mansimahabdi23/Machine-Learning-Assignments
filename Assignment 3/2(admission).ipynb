```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Admission_Predict.csv")
df.head()
```

1. SCATTER PLOT

```python
plt.scatter(df["TOEFL Score"], df["GRE Score"], alpha=0.5)
plt.xlabel("TOEFL Score")
plt.ylabel("GRE Score")
plt.show()
```

2. BAR PLOT

```python



plt.bar(df["TOEFL Score"], df["GRE Score"], color='purple')
plt.xlabel("TOEFL Score")
plt.ylabel("GRE Score")
plt.title("TOEFL vs GRE Scores")
plt.show()
```

3. BOX PLOT

```python
import matplotlib.pyplot as plt

plt.boxplot([df["CGPA"]])

# Show the plot
plt.show()
```

4. PIE CHART

```python
# PIE CHART
labels = df['University Rating'].value_counts().index
sizes = df['University Rating'].value_counts().values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of University Ratings')
plt.show()
```

5. Line Chart

```python
plt.plot(df['TOEFL Score'], df['Chance of Admit'], marker='o')
plt.xlabel('TOEFL Score')
plt.ylabel('Chance of Admit')
plt.title('Chance of Admit over TOEFL Score')
plt.grid(True)
plt.show()
```
