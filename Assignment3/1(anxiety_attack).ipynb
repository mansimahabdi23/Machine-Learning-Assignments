```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

DATA VISUALIZATION 

1) SCATTER PLOT

```python
df.columns
```

```python
#1 age vs anxiety attack

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Age'], y=df['Severity of Anxiety Attack (1-10)'], hue=df['Gender'], palette='plasma')
plt.title('Age vs Anxiety Attack')
plt.xlabel('Age', fontsize=12, color='purple')
plt.ylabel('Severity of Anxiety Attack', fontsize=12, color='blue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Age'], y=df['Sleep Hours'], hue=df['Gender'], palette='viridis')
plt.title('Age vs Sleep Hours')
plt.xlabel('Age', fontsize=12, color='purple')
plt.ylabel('Sleep Hours', fontsize=12, color='blue')
plt.show()
```

2) BAR PLOT

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Age'], y=df['Heart Rate (bpm during attack)'], palette='magma')
plt.title('Age vs Heart Rate')
plt.xlabel('Age', fontsize=12, color='purple')
plt.ylabel('Heart Rate', fontsize=12, color='blue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Physical Activity (hrs/week)'], y=df['Stress Level (1-10)'], palette='magma')
plt.title('Physical Activity vs Stress Level')
plt.xlabel('Physical Activity', fontsize=12, color='purple')
plt.ylabel('Stress Level', fontsize=12, color='blue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Occupation'], y=df['Sleep Hours'], palette='magma',hue=df['Gender'])
plt.title('Occupation vs Sleep Hours')
plt.xlabel('Occupation', fontsize=12, color='purple')
plt.ylabel('Sleep Hours', fontsize=12, color='blue')
plt.show()
```

3) BOX PLOT

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Occupation'], y=df['Stress Level (1-10)'], palette='magma')
plt.title('Occupation vs Stress Level')
plt.xlabel('Occupation', fontsize=12, color='purple')
plt.ylabel('Stress Level', fontsize=12, color='blue')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Gender', y='Severity of Anxiety Attack (1-10)', hue='Family History of Anxiety', data=df, palette='Blues',width=0.6, notch=True) 
plt.title('Severity of Anxiety Attack by Gender and Family History')
plt.show()
```

4) PIE CHART

```python
occupation_counts = df['Occupation'].value_counts()
plt.pie(occupation_counts,labels=occupation_counts.index,autopct='%1.1f%%',startangle=90,colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2', '#ff6666', '#c2f0f0', '#ffcc00'])
plt.axis('equal')
plt.title('Occupation Distribution')
plt.show()
```

```python
stress_level_counts = df['Stress Level (1-10)'].value_counts()
plt.pie(stress_level_counts,labels=stress_level_counts.index,autopct='%1.1f%%',startangle=90,colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2', '#ff6666', '#c2f0f0', '#ffcc00'])
plt.axis('equal')
plt.title('Stress Level Distribution')
plt.show()
```

5) LINE CHART

```python
import pandas as pd
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Age')

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(df_sorted['Age'], df_sorted['Stress Level (1-10)'], marker='o', linestyle='-', color='blue', linewidth=2, markersize=5, alpha=0.7, label="Stress Level")

# Customizations
plt.xlabel('Age')
plt.ylabel('Stress Level')
plt.title('Age vs Stress Level')
#plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (Use your actual CSV data)
# df = pd.read_csv('your_file.csv')

# Sorting by Caffeine Intake to smooth the line
df_sorted = df.sort_values(by='Caffeine Intake (mg/day)')

# Set Seaborn Style
sns.set_style("darkgrid")

# Plotting with Seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(data=df_sorted, x='Caffeine Intake (mg/day)', y='Heart Rate (bpm during attack)', marker='o', linewidth=2, markersize=6, color='blue')

# Customizations
plt.xlabel('Caffeine Intake (mg/day)', fontsize=12)
plt.ylabel('Heart Rate (bpm)', fontsize=12)
plt.title('Caffeine Intake vs Heart Rate', fontsize=14)
plt.xticks(rotation=30)  # Rotate X-axis labels for clarity
plt.show()
```
